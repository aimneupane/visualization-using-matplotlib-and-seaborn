# Matplotlib/Seaborn Strip and swarm plot:

* These both plots are absence in matplotlib only they are native to seaborn only.
* you can use it using scatter with the use of jitter but more complex and inconvenient.
* A strip plot shows all individual data points of a numerical variable across categories, spread slightly to avoid overlap.
* What it shows
    * Exact distribution of data
    * Outliers
    * Data density
* A swarm plot is like a strip plot, but:Points are automatically arranged so they never overlap and are used for medium datasets

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
tip=sns.load_dataset("tips")
tip
```

```python
# we normally use scatter plot if both data are numerical in nature.like
sns.scatterplot(data=tip,x='total_bill',y='tip')
```

but if we want to use categorical and numerical data in one frame and we want to analyze numerical data using categories then in that sense we use categorical plot ie. strip plot

```python
sns.catplot(data=tip,x='sex',y='total_bill',kind='strip') # figure level.
```

```python
sns.stripplot(data=tip,x='sex',y='total_bill') # axes level plot.
```

```python
# now see we are looking data to be in one line but the data are distributed significantly.
sns.stripplot(data=tip,x='sex',y='total_bill',jitter=False)
```

now the datas are aligned in one line position. and shows where all the numerical data lie in the graph.

#### Swarm plot

```python
sns.catplot(data=tip,x='sex',y='total_bill',kind='swarm') # figure level.
```

```python
sns.swarmplot(data=tip,x='sex',y='total_bill')  # axes level.
```

```python
sns.catplot(data=tip,x='sex',y='total_bill',kind='swarm') 
```

```python
sns.catplot(data=tip,x='day',y='total_bill',kind='swarm') 
```

Step-by-step: How to study a strip/swarm plot
Step 1: Compare vertical position
    * Higher dot: higher value
    * Lower dot : lower value
This tells magnitude

Step 2: Look at density of dots
    * Ask yourself: Where are most dots clustered? Are dots concentrated at top or bottom?
This tells typical range

Step 3: Look for spread
    * Tall vertical spread : high variability
    * Short vertical spread: low variability
This tells consistency of data

Step 4: Identify outliers
    * Dots far away from the main cluster
    * One or two dots extremely high or low
This tells unusual behavior

Step 5: Compare categories Ask yourself:
        * Which category has generally higher dots? Which category has wider spread? Which category has more extreme values?

From the swarm plot, it is observed that total bills on Saturday are generally higher and more variable compared to other days, with several high-value outliers.
