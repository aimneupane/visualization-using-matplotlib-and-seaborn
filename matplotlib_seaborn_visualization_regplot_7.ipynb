# Matplotlib Seaborn Visualization for regression plot:

Relationship between two numerical variables + a fitted regression line mostly used in prediction/machine learning scenario.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=sns.load_dataset("tips")
df
```

```python
# using seaborn:
sns.regplot(data=df,x='total_bill',y='tip')
```

there is relationship as total bill increases tips also increases showing positive correlation among them. and shaded part is confidence interval.

use seaborn in this instead of matplotlib but if you want to learn real math problem them you can use matplotlib but needs extra more code.

```python
x = df['total_bill']
y = df['tip']

# scatter plot
plt.scatter(x, y)

# regression calculation
m, c = np.polyfit(x, y, 1)   # slope & intercept
y_pred = m*x + c

# regression line
plt.plot(x, y_pred)

plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Regression using Matplotlib")
plt.show()
```
