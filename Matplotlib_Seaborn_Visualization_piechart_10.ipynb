```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
df=sns.load_dataset("tips")
df
```

```python
# pie chart is only done with the help of matplotlib. it is absent in seaborn library.
count=df['sex'].value_counts()
count
```

```python
plt.pie(count.values,labels=count.index,autopct='%.1f%%',explode=[0.1,0],shadow=True)
plt.show()
```

pie chart shows how much each labels is contributing on our datasets or any given mathematical form.
