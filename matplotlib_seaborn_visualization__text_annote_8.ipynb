# Matplotlib Seaborn Annotation putting text:

Adding text labels at specific points in a graph to explain, highlight, or describe important information.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
df=sns.load_dataset("tips")
df
```

#### As this is not the seaborn inbuilt features it builts on top of matplotlib. so, either you use any language matplotlib annotation is needy.

```python
plt.scatter(df['total_bill'],df['tip'])
```

now put annotation in it.

```python
# on scatter plot:
plt.scatter(df['total_bill'],df['tip'])
plt.annotate(
        "max point",
        xy=(50,10), # this means x on 50 and y on 10.
        xytext=(30,8), # this means the text is written in this coordinate.
        arrowprops=dict(arrowstyle="->",color='r')
        )
```

```python
# on line plot:
sns.lineplot(data=df,x='total_bill',y='tip')
plt.annotate(
            "highest tip",
            xy=(49,9),
            xytext=(20,8),
            arrowprops=dict(arrowstyle='->',color='r')
            )
```

```python
# on bar plot:
sns.barplot(data=df,x='day',y='total_bill')
plt.annotate(
"Max",
xy=(3,21.5),
xytext=(1,24),
    arrowprops=dict(arrowstyle='->',color='r')
)
```

```python
# on bar plot:
sns.barplot(data=df,x='day',y='total_bill')
plt.annotate(
"Max",
xy=(3,21.5),
xytext=(1,24),
    arrowprops=dict(arrowstyle='->',color='r')
)
plt.annotate(
"Min",
xy=(1,17),
xytext=(0,23),
    arrowprops=dict(arrowstyle='->',color='r')
)
```

```python
# dynamic annotation:
ax=sns.barplot(data=df,x='day',y='total_bill')
# Get the sunday bar:
sunday_bar=ax.patches[3]
# Get the friday bar:
friday_bar=ax.patches[1]

x=sunday_bar.get_x()+sunday_bar.get_width()/2
y=sunday_bar.get_height()

xx=friday_bar.get_x()+friday_bar.get_width()/2
yy=friday_bar.get_height()

ax.annotate(
"max",
xy=(x,y),
xytext=(x-1,y+2),
arrowprops=dict(arrowstyle='->',color='r'),
ha='right'
)
ax.annotate(
"min",
xy=(xx,yy),
xytext=(xx-0.5,yy+2),
arrowprops=dict(arrowstyle='->',color='r'),
)
plt.show()
```

```python
# Create bar plot (total bill per day)
ax = sns.barplot(data=df, x='day', y='total_bill')

# Add value on top of each bar
for bar in ax.patches:
    x = bar.get_x() + bar.get_width() / 2   # center of bar
    y = bar.get_height()                     # top of bar
    ax.text(x, y + 0.5, f"{y:.1f}", ha='right')  # 0.5 moves text slightly above

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 20]

plt.plot(x, y, marker='o')

for xi, yi in zip(x, y):  # zip iterates/combines multiple list simultaneously.
    plt.text(xi, yi + 0.5, str(yi), ha='center')
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 20]

# First, plot the line or points
plt.plot(x, y, marker='o')  # or plt.scatter(x, y)

# Now add text labels
for i in range(len(x)):
    plt.text(x[i], y[i] + 0.5, str(y[i]), ha='center')

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 20]
demo=pd.DataFrame({
    'X':x,
    'Y':y,
    'Z':['Ram','Shyam','Hari','Karma','Bijay']
})
demo
```

```python
plt.scatter(demo['X'],demo['Y'])
for i in range(demo.shape[0]):
    plt.text(demo['X'].values[i],demo['Y'].values[i],demo['Z'].values[i])
```
