# Matplot Seaborn visualization for Bar plot

* A bar graph (bar chart) is a categorical plot that uses rectangular bars to represent numerical values for different categories.
Why do we need a Bar Graph?
    * We need a bar graph to:
    * Compare values between different categories
    * Show counts, sums, means, or percentages
    * Quickly identify highest and lowest categories

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
df=sns.load_dataset("tips")
df
```

```python
# using matplotlib:
plt.bar(df['smoker'],df['total_bill'])
```

```python
plt.bar(df['sex'],df['total_bill'])
```

```python
# where do we need bar graph in horizontal orientation if categories are many then we have to use.
# lets use different dataset.
data=px.data.gapminder()
data
```

```python
data['country'].unique()
```

```python
plt.bar(data['country'],data['pop'])
```

```python
data=data[data["country"].isin(["Nepal","India","China","Bangladesh","Pakistan","Australia","Zimbabwe","Portugal","United States"])]
```

```python
data
```

```python
# Look the bar graph is distorted as we can't see all the data in clear format.
plt.barh(data['country'],data['pop'],color=['r','g','b','y','r','g','b','y','b'])
```

```python
top10 = data.sort_values(by='pop', ascending=False).head(50)

plt.barh(top10['country'], top10['pop'])
plt.xlabel("Population")
plt.title("Top 10 Countries by Population")
plt.show()
```

```python
# can i adjust the width of the bar:? yes, ofcourse.
plt.bar(top10['country'], top10['pop'],width=0.5)
plt.xlabel("Population")
plt.title("Top 10 Countries by Population")
plt.show()
```

```python
# can i get errorbars in my bar too: let 
top10 = data.sort_values(by='pop', ascending=False).head(10)
error = [10000, 20000, 50000, 35000, 40000,25000, 30000, 15000, 28000, 22000]
error_million = np.array(error) / 1e6
# can i adjust the width of the bar:? yes, ofcourse.
plt.bar(top10['country'], top10['pop'],yerr=error_million,capsize=5,width=0.5)
plt.xlabel("Population")
plt.title("Top 10 Countries by Population")
plt.show()
```

```python
# how to use multiple bar graph:lets use new dataset.
df=pd.read_csv("batsman_season_record.csv")
df
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

```python
x=np.arange(len(df['batsman']))#here use numpy
w=0.4
plt.bar(x-w/2,df['2015'],w)
plt.bar(x,df['2016'],w)
plt.bar(x+w/2,df['2017'],w)
```

```python
# it is looking good but x-axis is not showing relevant batsman. so solve this.
plt.bar(x-w/2,df['2015'],w,label='2015')
plt.bar(x,df['2016'],w,label='2016')
plt.bar(x+w/2,df['2017'],w,label='2017')
plt.xlabel("batsman")
plt.ylabel("runs scored")
plt.legend()
plt.xticks(x-w/4,df['batsman'].values)
plt.show()
```

```python
# use stacked bar graph also:
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

```python
# this stacked is not good. as not showing all the details.
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'],bottom=df['2015'])
plt.bar(df['batsman'],df['2017'],bottom=df['2015']+df['2016'])
```

```python
sns.catplot(data=df,x='batsman',y='2015',kind='bar') # figure level:
```

```python
sns.barplot(data=df,x='batsman',y='2015') #axes level plot:
```
