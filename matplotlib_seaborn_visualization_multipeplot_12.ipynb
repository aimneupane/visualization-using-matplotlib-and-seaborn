# Matplotlib Seaborn Visualization Multi Plots:

Multi-plots answer many related questions at once

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
df=px.data.gapminder()
df
```

### using subplot:

```python
nepal=df[df['country']=='Nepal']
nepal
```

```python
plt.subplot(1,2,1)
plt.plot(nepal['year'],nepal['lifeExp'])
plt.title("Life Expectancy")

plt.subplot(1,2,2)
plt.plot(nepal['year'],nepal['gdpPercap'])
plt.title("GDP")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)
ax = ax.flatten()

ax[0].plot([1,2,3], [4,5,6])
ax[1].plot([1,2,3], [6,5,4])
ax[2].plot([1,2,3], [6,5,4])
ax[3].plot([1,2,3], [6,5,4])

plt.show()
```

### using grid:

```python
fig,ax=plt.subplots(2,2,figsize=(8,6))

ax[0,0].plot(nepal['year'],nepal['lifeExp'])
ax[0,0].set_title("Life Expectancy trend compare")

ax[0,1].plot(nepal['year'],nepal['gdpPercap'])
ax[0,1].set_title("GDP per capita trend compare")

ax[1,0].scatter(nepal['year'],nepal['lifeExp'])
ax[1,0].set_title("Life Expectancy relationship")

ax[1,1].scatter(nepal['year'],nepal['gdpPercap'])
ax[1,1].set_title("GDP per capita relationship")

fig.text(0.5, -0.05, "Figure: Demonstration of subplots",
         ha='center')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)

for axes in ax.flat:
    axes.plot([1,2,3], [1,4,9])
    axes.scatter([1,2,3], [1,4,9])

plt.show()
```

```python
fig, ax = plt.subplots(2, 2)

for i, axes in enumerate(ax.flat):
    if i % 2 == 0:
        axes.plot([1,2,3], [1,4,9])      # line
    else:
        axes.scatter([1,2,3], [1,4,9])   # scatter

plt.show()
```

### seaborn:

### pairplot:
* explore all numerical relationships.
* in diagonal distribution is shown with kde or hist plot.
* in non-diagonal shows with scatter plot.

```python
tip=sns.load_dataset("tips")
tip
```

```python
sns.pairplot(data=tip)
```

```python
sns.pairplot(data=tip,hue='sex')
```

### joint plot:
* it shows relationship with distribution. needy for deep diving into two variables.

```python
sns.jointplot(data=tip,x='total_bill',y='tip',kind='scatter')
```

### Facet:

```python
g=sns.FacetGrid(data=tip,col='sex')
g.map(sns.histplot,'total_bill')
```

```python
sns.FacetGrid(tip, col="sex") \
   .map(sns.histplot, "total_bill")
```

```python
g=sns.FacetGrid(tip, row="sex", col="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
```

```python
g=sns.FacetGrid(tip, row="sex", col="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
```

```python
sns.relplot(
    data=tip,
    x="total_bill", y="tip",
    col="sex", row="smoker",
    kind="scatter"
)
```

```python
sns.relplot(
    data=tip,
    x="total_bill", y="tip",
    col="day",
    col_wrap=2,
    kind="scatter"
)
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

fig = plt.figure(figsize=(8,6))

# Create a 2x2 grid
gs = GridSpec(2, 2, figure=fig)

# Top-left plot
ax1 = fig.add_subplot(gs[0, 0])
ax1.plot(x, y1)
ax1.set_title("Sine")

# Top-right plot spanning two rows
ax2 = fig.add_subplot(gs[:, 1])
ax2.plot(x, y2, color='orange')
ax2.set_title("Cosine")

# Bottom-left plot
ax3 = fig.add_subplot(gs[1, 0])
ax3.plot(x, y3, color='green')
ax3.set_title("Tangent")

plt.tight_layout()
plt.show()
```
