# Matplotlib Seaborn Scatter Plot Visualization

##### A scatter plot is used to show the relationship, pattern, or correlation between two numerical variables.
* When to use scatter plot
    * Both variables are numerical
    * You want to check association
    * Before doing regression analysis
    * positive or negative trend
    * strength of correlation
    * Data spreading and clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
df=sns.load_dataset("tips")
df
```

```python
# using matplotlib:
plt.scatter(df['total_bill'],df['tip'])
```

```python
plt.scatter(df['total_bill'],df['tip'],c=df['size'],s=df['size']*20,alpha=0.5,label='tip')
plt.title("tip vs totalbill")
plt.xlabel("total_bill")
plt.ylabel("tips")
plt.legend()
plt.xlim(0, 80)   # range
plt.xticks([1,5,10,15,20,25,30,35,40,45,50,60,70,80])
plt.show()
```

```python
# using seaborn:
sns.relplot(data=df,x='total_bill',y='tip',kind='scatter')  # figure level plot.
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip',hue='sex',style='smoker',size=df['size'])  # axes level plot.
```

### lets discuss why matplotlib is important while plotting?

```python
bat=pd.read_csv("batter.csv")
bat
```

```python
filter_batter=bat.head(100).sample(25,random_state=5)
filter_batter
```

```python
# now plot this.
plt.scatter(filter_batter['strike_rate'],filter_batter['avg'])
```

```python
# but we donot know whose points are these? Can you guess? it's really hard. so, we need some coding or annotate.
```

```python
plt.figure(figsize=(10,6))
plt.scatter(filter_batter['avg'],filter_batter['strike_rate'])
plt.axvline(30,color='red')
plt.axhline(130,color='green')
for i in range(filter_batter.shape[0]):
    plt.text(filter_batter['avg'].values[i],filter_batter['strike_rate'].values[i],filter_batter['batter'].values[i])
```
