# Matplotlib and Seaborn Box Plot and Violin Plot:

### Box plot: Why do we need boxplots?
* Bar plots hide variation
* Strip/swarm plots get cluttered for large data
* Boxplots give fast comparison
##When should you use boxplots?Conditions:
* One numerical variable
* One categorical variable (optional)
* Medium to large dataset
##How to read a boxplot
* Line inside box: Median
* Box:middle 50% of data (median)
* Whiskers:data range
* Points outside: Outliers

### Violin Plot: A violin plot combines:
* Boxplot statistics
* Kernel Density Estimation (distribution shape)
* It shows how data is distributed, not just summary values.
##What does a violin plot show?
* Distribution shape
* Density of values
* Median & quartiles (optional)
* Multiple peaks (bimodal data)
##Why do we need violin plots?
* Boxplots cannot show:
* Multiple peaks
* Distribution shape
##Violin plots can.When should you use violin plots?Conditions:
* Numerical variable with many observations
* Want to see distribution shape
* Comparing multiple categories

```python
# both can be done with matplotlib and seaborn.
# but it is recommended to use seaborn for both for better interactivity.
# while violin plot has less features in matplotlib so, better to use seaborn in both cases.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
tip=sns.load_dataset("tips")
tip
```

```python
# using boxplot with matplotlib:
plt.boxplot(tip['total_bill'],vert=True,patch_artist=True)
plt.show()
```

```python
# using seaborn:
sns.catplot(data=tip,x='sex',y='total_bill',kind='box') # figure level and more interactiveness.
```

```python
sns.boxplot(data=tip,x='sex',y='total_bill') # axes level
```

```python
sns.boxplot(data=tip,x='sex',y='total_bill',hue='smoker') # axes level
```

```python
# from the figure upper whisker is maximum value, lower whisker is minimum value.
# upper part of the box is 3rd Quartile and lower part is first Quartile while middle line is second quartile or median.
# before and after the whisker the points are said to be outliers. we can see outliers in all boxes.
```

#### use of violin plot:

```python
sns.catplot(data=tip,x='sex',y='total_bill',kind='violin') # figure  level
```

```python
sns.violinplot(data=tip,x='sex',y='total_bill') # axes level
```

```python
sns.violinplot(data=tip,x='sex',y='total_bill')
```

```python
# from the figure we can see inside the violin like shaped there is box plot with whisker which is shown with black colour.
# and the kde plot is placed outside the box plot which shows distribution in bi-modal nature.
```

```python
sns.violinplot(data=tip,x='sex',y='total_bill',inner='box')
```
