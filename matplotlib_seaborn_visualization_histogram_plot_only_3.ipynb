# Histogram Plot using matplotlib and seaborn:

* A histogram is a graphical representation of the distribution of numerical data, where data is grouped into bins (intervals) and displayed as bars.
* Why do we need a Histogram?
    * Because raw numbers do not show patterns easily.
    * A histogram helps us to:
    * Understand data distribution
    * See frequency of values
    * Identify skewness
    * Detect outliers
    * Analyze spread of data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
df=sns.load_dataset("tips")
df
```

```python
# using matplotlib:
plt.hist(df['total_bill'])
plt.title("frequency distribution of total bill")
plt.xlabel("total bill")
plt.ylabel("frequency")
plt.show()
```

```python
# can we adjust bins and can we use extra parameters in this:?  yes, absolutely.
```

```python
plt.hist(df['total_bill'],bins='auto',color='g',edgecolor='black',alpha=0.4,density=True) #Y-axis becomes probability density normalization.
plt.title("frequency distribution of total bill")
plt.xlabel("total bill")
plt.ylabel("frequency")
plt.show()
```

```python
# can we limit the range of x-axis:
plt.hist(df['total_bill'],bins='auto',color='g',edgecolor='black',range=(0,70))
plt.title("frequency distribution of total bill")
plt.xlabel("total bill")
plt.ylabel("frequency")
plt.show()
```

```python
# bar style:
plt.hist(df['total_bill'],bins='auto',color='g',edgecolor='black',range=(0,70),histtype='step')
plt.title("frequency distribution of total bill")
plt.xlabel("total bill")
plt.ylabel("frequency")
plt.show()
```

```python
# can the orientation be changed:
plt.hist(df['total_bill'],bins='auto',color='g',edgecolor='black',range=(0,70),histtype='step',orientation='horizontal')
plt.title("frequency distribution of total bill")
plt.xlabel("total bill")
plt.ylabel("frequency")
plt.show()
```

```python
# if the data are so big. to analyze them we have to make into small.so we have to use logarithmic scale.
plt.hist(df['total_bill'],bins='auto',color='g',edgecolor='black',log=True)
plt.title("frequency distribution of total bill")
plt.xlabel("total bill")
plt.ylabel("frequency")
plt.show()
```

```python
plt.hist(df['total_bill'],bins='auto',color='g',edgecolor='black',cumulative=-1)
plt.title("frequency distribution of total bill")
plt.xlabel("total bill")
plt.ylabel("frequency")
plt.show()
```

```python
# can we use multiple hist in one frame:
```

```python
plt.hist(df[['total_bill','tip']],label=['A','B'])
plt.legend()
```

```python
plt.hist(df[df['sex']=='Male']['total_bill'], bins=10, alpha=0.5, label='Male')
plt.hist(df[df['sex']=='Female']['total_bill'], bins=10, alpha=0.5, label='Female')
plt.legend()
plt.show()
```

```python
# using seaborn:
sns.displot(df['total_bill'])  # figure level plot
```

```python
sns.histplot(df['total_bill'])  # axes level plot
```

```python
# can we use other parameters in this? yes , absolutely.
sns.displot(data=df,x='total_bill',color='g',hue='sex',kde=True,multiple='stack')  # figure level plot
```
