# Matplotlib and Seaborn lineplot only

### Differences between them lets dive into it:
    * Matplotlib is a low level python library used for creating static, interactive and animated plots.
    * While Seaborn is a high level visualization library built on top of matplotlib.
_______------------------_______________

* Matplotlib gives full control over plots
* Why Matplotlib?
    ** Full control
    ** Base for almost all plotting libraries
    ** Needed for customization
    ** Required for animations
* Why Seaborn?
    ** Faster visualization
    ** Beautiful default styles
    ** Ideal for data analysis & statistics

ðŸ§  Seaborn cannot work without Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px  # used just for the dataset
%matplotlib inline
```

```python
# lets extract dataset using 
sns.get_dataset_names()
```

```python
data=px.data.gapminder()
data
```

```python
data['country'].unique()
```

```python
df=data[data['country']=="Nepal"]  # we are extracting for Nepal only.
df
```

### First of all we are going to plot for line plot only in this class. we will be goind deep on this using matplotlib and seaborn
* using matplotlib we have to use this  plt.plot(x-coordinate,y-coordinate)
* for seaborn we have to use:
    * for figure level axes: sns.relplot(data=dataset, x="", y="", kind="line")
    * for axes level: sns.lineplot(data=dataset, x="", y="")
    * in seaborn we have option to use hue attribute, style and size attribute for better visualization.

```python
# using matplotlib:
plt.plot(df['year'],df['lifeExp'])
```

```python
font={'size':12,'color':'g'}
plt.plot(df['year'],df['lifeExp'],label="Nepal",marker='o',linestyle='dashed',mec='g',mfc='y',linewidth=2)
plt.xlabel("year",fontdict=font)
plt.ylabel("life Expectancy",fontdict=font)
plt.title("Nepal life Expectancy upto 2007",fontdict=font)
plt.legend()
plt.show()

# here label is written and it's for legend.
# marker is o linestyle is dashed line. mec means marker edge color while mfc is marker face color.
# linewidth is 2
```

```python
# using seaborn:
sns.relplot(data=df,x='year',y='lifeExp',kind='line',color='g') # figure level axes.
```

```python
sns.lineplot(data=df,x='year',y='lifeExp') # for axes level.
```
